```python
# Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
pd.read_csv()
```

```python
# read the wine quality dataset
df = pd.read_csv("/content/WineQT (1).csv").drop(columns = ['Id'], axis  =1)
df.head()
```

```python
df.tail()
```

```python
# Check the unique values of Wine Quality
df['quality'].unique()
```

```python
# Create a function to convert the quality into Good (1) or Bad (0)
def good_or_bad(val):
    if val > 6:
        return 1
    else:
        return 0

df['quality'] = df['quality'].apply(good_or_bad)
df.head()
```

```python
df.tail()
```

```python
# Create X and y Variables
X = df.drop(columns=['quality'])
y = df['quality']

# Create the MinMaxScaler instance and convert the X values into Normalized Dataframe
X_norm = MinMaxScaler().fit_transform(X)
X_norm_df = pd.DataFrame(X_norm, columns=df.columns.to_list()[:-1])
X_norm_df.head()
```
